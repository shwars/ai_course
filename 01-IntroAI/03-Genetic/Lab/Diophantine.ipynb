## Лабораторная работа: Диофантовы уравнения

Цель работы - научиться решать **Диофантовы уравнения** - линейные уравнения с целыми корнями и целыми коэффициентами. Например, рассмотрим следующее уравнение:

$$a+2b+3c+4d=30$$

Необходимо найти целочисленные корни $a$,$b$,$c$,$d\in\mathbb{N}$, удовлетворяющие этому уравнению.

Подсказки:
1. Можно считать, что корни принадлежат некоторому интервалу, например, [0;30]
1. В качестве гена попробуйте использовать список потенциальных корней
1. Подумайте, что использовать в качестве фитнес-функции.
